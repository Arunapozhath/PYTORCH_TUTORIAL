# In Convolutional Neural Networks (CNNs), **image filters (also called kernels**)

are the core components that extract features from input images. These filters help the network learn spatial hierarchies of patterns, from simple edges to complex shapes.** Let's break down how filters work in CNNs:**

1. What is a Filter (Kernel) in CNN?
A filter or kernel is a small matrix (e.g., 3x3, 5x5) that is applied (convolves) over the input image or the feature maps to detect features like edges, textures, and patterns.
The filter moves across the image, performing element-wise multiplication with the pixel values in the image, and summing the results to produce a new output value (this process is called a convolution).
For example, a 3x3 filter may look like this:


[[-1, -1, -1],

[ 0,  0,  0],

[ 1,  1,  1]]

This kernel might be used to detect horizontal edges in an image.

2. How Filters Work in CNNs?
The filters in CNNs extract various features from the image through the following process:

**Step 1: Convolution Operation**

The filter slides (or convolves) over the input image, at each position multiplying the filter values with the image pixel values, then summing the result.

This produces a feature map, which contains the transformed representation of the image with respect to the features that the filter detects.

**Step 2: Multiple Filters**

CNNs use multiple filters in each convolutional layer. Each filter is responsible for extracting a specific feature from the image. For example:
Edge detection filters may capture vertical, horizontal, or diagonal edges.
Texture detection filters might capture patterns like ripples or spots.
Object shape detection filters, in deeper layers, capture more complex features such as parts of objects.

**Step 3: Feature Maps**

The output of applying a filter on the input image is called a feature map. Multiple filters applied to the same image produce multiple feature maps, each highlighting different aspects of the image.

# ***CNN Structure and RGB 3D in Image Processing***

A **Convolutional Neural Network (CNN)** is a type of deep learning model particularly well-suited for processing **grid-like data**, such as images. The network's architecture is designed to capture spatial hierarchies by applying filters (or kernels) to an input image. Here's how CNN structure works, especially in the context of **RGB images**, which are 3D data.

### 1. **Basic CNN Structure**

The architecture of a CNN is typically composed of the following layers:

#### a. **Input Layer**
- **Input Data**: For CNNs, the input is usually an image. In the case of an **RGB image**, the input is a 3D array (height, width, channels).
  - An RGB image has 3 channels: **Red**, **Green**, and **Blue**.
  - For example, if an RGB image is 224x224 pixels, the input shape will be **224x224x3** (height x width x channels).
  
#### b. **Convolutional Layer**
- The **convolutional layer** is where the CNN applies multiple filters to the input data.
  - Each filter is a small 2D or 3D matrix (e.g., 3x3x3 for RGB images) that convolves over the image to detect features.
  - The output of this operation is called a **feature map**, which is a 2D matrix representing detected features like edges or textures in the image.
  - Multiple filters are applied, creating multiple feature maps.

#### c. **Activation Layer (ReLU)**
- **ReLU (Rectified Linear Unit)** is the most common activation function used in CNNs.
  - It introduces **non-linearity** by setting all negative values in the feature map to zero while keeping positive values unchanged.
  - This helps the CNN model learn complex patterns.

#### d. **Pooling Layer**
- **Pooling** layers reduce the spatial dimensions (height and width) of the feature maps, which helps make the computation more efficient.
  - Commonly used pooling is **max pooling**, where only the maximum value from each small region of the feature map is retained.
  - Pooling helps the CNN become more invariant to small translations of the input image (e.g., if the object moves slightly).

#### e. **Fully Connected Layer**
- After multiple convolution and pooling layers, the final layers are usually **fully connected**.
  - Here, the feature maps are flattened into a 1D vector, and a traditional neural network layer is applied.
  - These layers are used for the final classification or prediction tasks (e.g., determining what object is in an image).

#### f. **Output Layer**
- The **output layer** depends on the task:
  - For **classification**, it might be a softmax layer that outputs probabilities for each class.
  - For **regression**, it might be a linear layer that outputs a continuous value.

### 2. **RGB Images and 3D Input (Three Channels)**

#### a. **3D Nature of RGB Images**
- An RGB image has three color channels: **Red**, **Green**, and **Blue**.
  - Each pixel in an RGB image is represented by a **3D vector** where each dimension corresponds to the intensity of the red, green, and blue components.
  - For example, a single pixel in an RGB image could be represented as **[R, G, B]**, where the values range from 0 to 255, indicating the intensity of each color.

#### b. **3D Input to CNN**
- The input to a CNN for an RGB image is a **3D array**, where the dimensions are:
  - **Height** (number of rows of pixels)
  - **Width** (number of columns of pixels)
  - **Channels** (3 channels for Red, Green, and Blue)
  
For example:
- An image of size 224x224 pixels will have the shape **224x224x3**.
  
#### c. **Convolutions on RGB Images**
- Filters in the convolutional layers also take into account the 3 channels of the RGB image. For example:
  - A **3x3x3 filter** will have dimensions that match the width and height (3x3) and will also span the 3 color channels (Red, Green, and Blue).
  
- This means that convolution operations will compute feature maps by summing over the three channels, effectively considering how color information in different channels interacts.

#### d. **Feature Maps in RGB Images**
- After the convolution operation, the **output feature map** for an RGB image will have the shape of **H x W x N**, where:
  - **H**: Height of the feature map.
  - **W**: Width of the feature map.
  - **N**: Number of filters applied (the depth of the feature map).
  
  For example, if you apply 32 filters to a 224x224x3 image, the feature map output will be **224x224x32**, where each filter detects a specific feature from the RGB channels.

### 3. **Role of RGB in CNNs**
- **Color Information**: RGB images carry color information that can be crucial for distinguishing objects or patterns. For example, a CNN may learn that "green" pixels are likely part of a tree, while "blue" pixels could be part of the sky.
- **Spatial Relationships**: CNN filters learn spatial relationships between the three color channels, helping the model detect objects based on both color and spatial structure.

### 4. **Example of a CNN with RGB Input:**
Here's an example flow for a CNN that processes an RGB image:

1. **Input Layer**: The input image is 224x224x3.
2. **First Convolutional Layer**: Applies 64 filters of size 3x3x3.
   - The resulting feature map has a shape of 224x224x64.
3. **ReLU Activation**: Introduces non-linearity.
4. **Max Pooling Layer**: Reduces the size of the feature map, outputting 112x112x64.
5. **Second Convolutional Layer**: Applies 128 filters of size 3x3x64.
   - The resulting feature map has a shape of 112x112x128.
6. **Further Layers**: The process continues through additional convolution, activation, and pooling layers.
7. **Fully Connected Layer**: After flattening, fully connected layers process the high-level features for classification.
8. **Output Layer**: Outputs the final class probabilities (e.g., 10 classes for object recognition).

# ***POOLING LAYER***

A **Pooling Layer** in a **Convolutional Neural Network (CNN)** is like a tool that helps make the image smaller, simpler, and easier to process, without losing important information.

### How it works (a simple explanation):
Imagine you have a **big picture** made of little squares, and you want to keep the important parts while ignoring extra details. The pooling layer takes small parts of the picture (like 2x2 squares of pixels) and picks the most important value from each square.

### Two common types of pooling:
1. **Max Pooling**: It looks at each small section (like a 2x2 grid of numbers) and picks the **biggest number**. This helps the CNN focus on the strongest features (like a sharp edge or bright spot).
   - Example: From a square like [3, 1, 2, 5], max pooling picks **5** because it’s the biggest.

2. **Average Pooling**: It calculates the **average** of all the numbers in that small section.
   - Example: From a square like [3, 1, 2, 5], average pooling would give you **(3+1+2+5)/4 = 2.75**.

### Why is pooling important?
- **Makes images smaller**: It reduces the size of the image or feature map, which helps the CNN work faster and with less memory.
- **Keeps important features**: Even though the image gets smaller, pooling keeps the key information (like where the most important parts of the image are).
- **Makes the CNN better at recognizing things**: Pooling helps the network ignore small changes or shifts in the image, so it can focus on the bigger picture (like recognizing a cat whether it's on the left or right of the image).

### Example:
If you have a big picture of a cat, the pooling layer will help the CNN shrink it down while still knowing it’s a cat by focusing on the important features (like ears, eyes, and whiskers).

# ***MODEL***

class CNNModel(nn.Module): Defines a class CNNModel which inherits from nn.Module, the base class for all neural networks in PyTorch.

super().__init__(): Calls the parent class (nn.Module) constructor to initialize the model.

nn.Conv2d(...): Defines a 2D convolutional layer.
in_channels=1: The number of channels in the input image (1 channel for grayscale images).

out_channels=6: The number of output feature maps (or filters).

kernel_size=3: The size of the convolution kernel (3x3).

stride=1: The stride of the convolution (moves 1 pixel at a time).

padding=1: Adds 1 pixel of padding to the image to maintain the size.

The second convolution layer conv2 is similar but increases the out_channels to 16, meaning it will output 16 feature maps.

nn.Linear(...): Defines fully connected (FC) layers.

fc1: Takes the flattened input of size 7*7*16 (after the convolution and pooling layers) and maps it to 120 neurons.

fc2: Takes the 120 outputs from the first FC layer and maps them to 84 neurons.

fc3: The final FC layer maps the 84 outputs to 10 neurons (for a 10-class classification problem like MNIST digits 0–9).

Forward Pass: Defines how data flows through the model.

x = F.relu(self.conv1(x)): Applies the first convolution layer conv1 to input x, followed by a ReLU activation function to introduce non-linearity.

x = F.max_pool2d(x, 2, 2): Applies 2x2 max pooling to reduce the spatial dimensions of the feature maps (halves the height and width of the feature maps).

The same process is repeated with conv2, followed by another ReLU and max pooling.

x = x.view(-1, 7*7*16): This reshapes (flattens) the output from the previous layers into a 1D tensor so it can be passed to the fully connected layers.
The -1 means PyTorch automatically infers the batch size.

7*7*16 is the number of features after two max-pooling operations (28x28 → 14x14 → 7x7).

x = F.relu(self.fc1(x)): Applies the first fully connected layer (fc1), followed by ReLU activation.
x = F.relu(self.fc2(x)): Applies the second fully connected layer (fc2) with ReLU.

x = self.fc3(x): Outputs the logits from the final fully connected layer (fc3).

return F.log_softmax(x, dim=1): The final output is passed through the log_softmax function, which is typically used for multi-class classification. It converts the logits to log probabilities, making them ready for calculating loss with CrossEntropyLoss.


```python
class CNNModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(in_channels=6, out_channels=16, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(in_features=7*7*16, out_features=120)
        self.fc2 = nn.Linear(in_features=120, out_features=84)
        self.fc3 = nn.Linear(in_features=84, out_features=10)


        #forward function
    def forward(self,x): #Removed extra indentation here
        x = F.relu(self.conv1(x))
        x = F.max_pool2d(x, 2, 2)#kernal 2 stride 2
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2, 2)


#re-view to flatten it out
        x = x.view(-1, 7*7*16)
        #fully connected layers
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)

        return F.log_softmax(x, dim=1)
```

# ***TRAIN TEST & WORK ***

time.time(): This captures the starting time to measure how long the training takes.

epochs = 5: This sets the number of epochs (how many times the model will train on the full dataset).

train_losses, test_losses, train_correct, test_correct: Lists to store the loss values and number of correct predictions for both training and testing data across epochs.

 Training Loop

Training loop iterates over each batch of training data in train_loader. For each batch:

Forward pass: The model predicts y_pred based on input X_train.

Loss calculation: Cross-entropy loss is computed between y_pred and true labels y_train.

Correct predictions: The number of correctly predicted labels is calculated by comparing the model's predictions to the true labels.

Backpropagation: Gradients are calculated and weights are updated using the optimizer.

Progress printing: Every 600 batches, the epoch number, loss, and batch number are printed.

Tracking losses: The loss for each batch is saved in train_losses. Similarly, the number of correct predictions is tracked with train_correct.

3. Testing Loop (within the training loop)

esting loop runs after the training loop for each epoch.

No gradients: torch.no_grad() is used to prevent the computation of gradients, as we don't need to update weights during testing.

Predictions on test data: The model generates predictions for each batch of test data X_test.

Correct predictions: The number of correct predictions for the test set is calculated and accumulated in tst_corr.

Loss calculation: Loss for the test set is computed and added to test_losses.

Results printing: At the end of each epoch, the training and testing losses and correct predictions are printed.

After the entire training process (across all epochs), the total time taken is calculated and printed in minutes.


```python
import time
start_time = time.time()

epochs = 5
train_losses = []
test_losses = []
train_correct = []
test_correct = []


#train
for i in range(epochs):
    trn_corr = 0  # Initialize correct predictions counter for this epoch
    tst_corr = 0  # Initialize correct predictions counter for test phase

    for b, (X_train, y_train) in enumerate(train_loader):
        b+=1  # Adjust batch count to start at 1 instead of 0
        y_pred = model(X_train)  # Forward pass: get model predictions
        loss = criterion(y_pred, y_train)  # Compute the loss between predictions and ground truth

        # Calculate how many predictions were correct
        predicted = torch.max(y_pred.data, 1)[1]  # Get the index of the highest logit for each prediction
        batch_corr = (predicted == y_train).sum()  # Compare predictions with the actual labels
        trn_corr += batch_corr  # Accumulate the number of correct predictions for this batch

        # Backpropagation and optimization step
        optimizer.zero_grad()  # Zero the gradients before backpropagation
        loss.backward()  # Backpropagate the loss
        optimizer.step()  # Update the model's weights

        # Print progress every 600 batches
        if b % 600 == 0:
            print(f'epoch: {i}  loss: {loss.item()}  b: {b}')

        # Append loss and correct predictions for this batch to tracking lists
        train_losses.append(loss.item())  # Save the loss for each batch
        train_correct.append(trn_corr.item())  # Save the number of correct predictions



#test

    with torch.no_grad():  # Disable gradient computation for testing
      for b, (X_test, y_test) in enumerate(test_loader):
        y_val = model(X_test)  # Forward pass: make predictions on the test data
        predicted = torch.max(y_val.data, 1)[1]  # Get the predicted class
        tst_corr += (predicted == y_test).sum()  # Count correct predictions

    # Compute loss for this test batch
    loss = criterion(y_val, y_test)
    test_losses.append(loss.item())  # Append test loss to the list
    test_correct.append(tst_corr.item())  # Append correct test predictions

    # Print test results after each epoch
    print(f'epoch: {i}  loss: {loss.item()}  b: {b}  train_corr: {trn_corr} test_corr: {tst_corr}')




current_time = time.time()
total = current_time - start_time
print (f'training took:{total/60}min')
```
